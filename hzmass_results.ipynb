# HyperzMass Results

This is where I'm going to clean and analyse the results from my HyperzMass runs. The four runs differed only in filter set:

1. all filters (i.e. u, g, r, i, z, VIS, Y, J, H), aka `all`
2. discounting the u-band (i.e. g, r, i, z, VIS, Y, J, H), aka `nou`
3. ground-based setup (i.e. u, g, r, i, z), aka `gnd`
4. euclid-only setup (i.e. VIS, Y, J, H), aka `euc`

These proved to be an interesting set of options to work with from our previous studies, and we hypothesise that the results from runs discounting the u-band may be surprisingly more accurate.

### Parameters

The parameters used in each run can be found under param in the GitHub repository **Link it here**. All the runs used the same version of Hyperz, the same templates, the same transmission filter.res file. The only change was to the filters 'mag' file, for which I commented out the filters that weren't to be included in each run. This also meant that I had to change the M_ABS filter choice, for which I chose the bluest filter in each run. 

The items selected from Claudia's simulation were identical for each run, the only difference in the catalogues was to not include the columns for filters we didn't want to include in the run.

#### Age

```python
plt.figure(figsize=(10,10))
plt.subplot(221)
plt.scatter(obs_df['t/Gyr'], all_zmass_df['age[Gyr]'], c=obs_df['Av'], s=5)
plt.xscale('log')
plt.yscale('log')

#plt.xlabel('log(M*/Msun)')
plt.subplot(222)
plt.scatter(obs_df['t/Gyr'], nou_zmass_df['age[Gyr]'], c=obs_df['Av'], s=5)
plt.xscale('log')
plt.yscale('log')
#plt.xlabel('log(Mgal/Msun)')
plt.subplot(223)
plt.scatter(obs_df['t/Gyr'], gnd_zmass_df['age[Gyr]'],  c=obs_df['Av'], s=5)
plt.xscale('log')
plt.yscale('log')
#plt.xlabel('SFR[Msun/yr]')
plt.subplot(224)
plt.scatter(obs_df['t/Gyr'], euc_zmass_df['age[Gyr]'],  c=obs_df['Av'], s=5)
plt.xscale('log')
plt.yscale('log')
#plt.xlabel('tau[Gyr]')
```

```python
all_age_acc = (all_zmass_df['age[Gyr]']- obs_df['t/Gyr'])#/(1 + obs_df['t/Gyr'])
nou_age_acc = (nou_zmass_df['age[Gyr]']- obs_df['t/Gyr'])#/(1 + obs_df['t/Gyr'])
gnd_age_acc = (gnd_zmass_df['age[Gyr]']- obs_df['t/Gyr'])#/(1 + obs_df['t/Gyr'])
euc_age_acc = (euc_zmass_df['age[Gyr]']- obs_df['t/Gyr'])#/(1 + obs_df['t/Gyr'])
```

```python
plt.figure(figsize=(10,10))
plt.suptitle('Age accuracy as a function of filter set (N=1000)')
plt.subplot(221)
sns.histplot(all_age_acc, binwidth=1, binrange=(-10, 15))
plt.xlabel(r'age$_{all}$ - age$_{true}$ [Gyr]')
plt.ylim(0, 600)

plt.subplot(222)
sns.histplot(nou_age_acc, binwidth=1, binrange=(-10, 15), color='r')
plt.xlabel(r'age$_{no-u}$ - age$_{true}$ [Gyr]')
plt.ylim(0, 600)

plt.subplot(223)
sns.histplot(gnd_age_acc, binwidth=1, binrange=(-10, 15), color='g')
plt.xlabel(r'age$_{gnd}$ - age$_{true}$ [Gyr]')
plt.ylim(0, 600)

plt.subplot(224)
sns.histplot(euc_age_acc, binwidth=1, binrange=(-10, 15), color='m')
plt.xlabel(r'age$_{euc}$ - age$_{true}$ [Gyr]')
plt.ylim(0, 600)

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.title('KDE of age accuracy as a function of filter set')
sns.kdeplot(all_age_acc, label='All bands')
sns.kdeplot(nou_age_acc, color='r', label='u-band discounted')
sns.kdeplot(gnd_age_acc, color='g', label='Ground-based bands')
sns.kdeplot(euc_age_acc, color='m', label='Euclid-only bands')
plt.legend()
plt.xlabel(r'age$_{hz}$ - age$_{true}$ [Gyr]')
plt.ylim(0, 0.5)

#plt.savefig('KDE_age.png', format='png')
```
